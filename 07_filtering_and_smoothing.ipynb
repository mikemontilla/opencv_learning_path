```python
import cv2
import numpy as np
from matplotlib import pyplot as pyplot

img_path = "images/thermal_image_2.png"
img = cv2.imread(img_path)

if img is None:
    print("Image not found in the provided path")
```

```python
# Mean blur
# larger kernel → stronger blur (k, k)
mean_blur_image_3= cv2.blur(img, (3, 3))
mean_blur_image_9= cv2.blur(img, (9, 9))
mean_blur_image_15= cv2.blur(img, (15, 15))

blur_path = "outputs/blur/"
cv2.imwrite(blur_path + "mean_blur_3.png", mean_blur_image_3)
cv2.imwrite(blur_path + "mean_blur_9.png", mean_blur_image_9)
cv2.imwrite(blur_path + "mean_blur_15.png", mean_blur_image_15)
```

```python
# gaussian blur
# kernel size must be odd: 3, 5, 7, 9, 11...
gaussian_blur_3 = cv2.GaussianBlur(img, (3, 3), 0)
gaussian_blur_9 = cv2.GaussianBlur(img, (9, 9), 0)
gaussian_blur_21 = cv2.GaussianBlur(img, (21, 21), 0)

cv2.imwrite(blur_path + "gaussian_blur_3.png", gaussian_blur_3)
cv2.imwrite(blur_path + "gaussian_blur_9.png", gaussian_blur_9)
cv2.imwrite(blur_path + "gaussian_blur_21.png", gaussian_blur_21)
```

```python
# median blur
# Notice the syntax difference
median_blur_3 = cv2.medianBlur(img, 3)
median_blur_5 = cv2.medianBlur(img, 5)
median_blur_9 = cv2.medianBlur(img, 9)

cv2.imwrite(blur_path + "median_blur_3.png", median_blur_3)
cv2.imwrite(blur_path + "median_blur_5.png", median_blur_5)
cv2.imwrite(blur_path + "median_blur_9.png", median_blur_9)

# Use median blur when you have noise like
# random white pixels
# random black pixels
# small isolated dots
```

```python
# Bilateral filter
# Bilateral filter smooths the image while trying to preserve edges
# d           → neighborhood diameter
# sigmaColor  → how different colors are mixed
# sigmaSpace  → how far pixels influence each other

bilateral = cv2.bilateralFilter(img, d=15, sigmaColor=75, sigmaSpace=75)
cv2.imwrite(blur_path + "bilateral.png", bilateral)
```
