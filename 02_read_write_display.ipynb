```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

image_path = "images/thermal_image_1.png"

img = cv2.imread(image_path)

if img is None:
    raise FileNotFoundError(f"It was not possible to find the file {image_path}")

print("Image loaded succesfully")

print("Shape: ", img.shape)
print("Dtype: ", img.dtype)
print("Height: ", img.shape[0])
print("Width: ", img.shape[1])
print("Number of channels: ", img.shape[2])
print("Minimum pixel value: ", np.min(img))
print("Maximum pixel value: ", np.max(img))


plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), vmin=0, vmax=255)
plt.axis('off')
plt.show()
```

```python
# exchange blue and red channels and save the image using different formats
image_path = "images/thermal_image_1.png"
thermal_image = cv2.imread(image_path)

if thermal_image is None:
    print("It was not possible to find the specified image")
else:
    print("Image loaded successfully")

exchanged_channels_image = thermal_image.copy()
exchanged_channels_image[:, :, 0] = thermal_image[:, :, 2]
exchanged_channels_image[:, :, 2] = thermal_image[:, :, 0]

plt.imshow(cv2.cvtColor(exchanged_channels_image, cv2.COLOR_BGR2RGB), vmin=0, vmax=255)
plt.axis('off')
plt.show()

# saving inverted image
os.makedirs("outputs", exist_ok=True)
saving_path = "outputs/channels_exchanged_thermal_image.jpg"
success = cv2.imwrite(saving_path, exchanged_channels_image)
success2 = cv2.imwrite(saving_path.replace("jpg", "png"), exchanged_channels_image)
success3 = cv2.imwrite(saving_path.replace("jpg", "bmp"), exchanged_channels_image)
print("Saved: ", success)
```

```python
# load in gray scale
image_path = "images/thermal_image_1.png"
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
if img is None:
    print("Image not found")
else:
    print("Image loaded successfully")

print("Shape: ", img.shape)

# saving image
saving_path = "outputs/gray_scale-thermal_image.jpg"
success = cv2.imwrite(saving_path, img)
print("Saved: ", success)

```

```python
# Convert to gray scale
image_path = "images/thermal_image_1.png"
img = cv2.imread(image_path)
if img is None:
    print("Image not found")
else:
    print("Image loaded successfully")

grays_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(grays_img.shape)

# saving image
saving_path = "outputs/gray_scale-thermal_image_converted.jpg"
success = cv2.imwrite(saving_path, grays_img)
print("Saved: ", success)
```

```python
# save a transformation
image_path = "images/thermal_image_1.png"
img = cv2.imread(image_path)
if img is None:
    print("Image not found")
else:
    print("Image loaded successfully")

small = cv2.resize(img, dsize=(0, 0), fx=0.5, fy=0.5)
crop = img[100:260, 400:900]
bright = cv2.convertScaleAbs(img, alpha=1.0, beta=40)

cv2.imwrite("outputs/thermal_small.jpg", small)
cv2.imwrite("outputs/thermal_crop.jpg", crop)
cv2.imwrite("outputs/thermal_bright.jpg", bright)
```
