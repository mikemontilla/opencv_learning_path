```python
import numpy as np

image = np.zeros((5, 5), dtype=np.uint8)
print("Shape: ", image.shape)
print("Dtype: ", image.dtype)
print(image)
```

```python
# white image
white_image = np.ones((10, 10), dtype=np.uint8) * 255

# change central pixel
white_image[5, 5] = 0 
print("Shape: ", white_image.shape)
print("Dtype: ", white_image.dtype)
print("Max value: ", np.max(white_image))
print("Min value: ", np.min(white_image))
print(white_image)
```

```python
# creating black image
height = 4
width = 6
one_pixel_colored_image = np.zeros((height, width, 3), dtype=np.uint8)

# change first pixel to blue
one_pixel_colored_image[0, 0] = [255, 0, 0]
print("Shape: ", one_pixel_colored_image.shape)
print("Dtype: ", one_pixel_colored_image.dtype)
print(one_pixel_colored_image)

```

```python
# indexing pixels

img = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
], dtype=np.uint8)

print("top-left pixel: ", img[0, 0])
print("center pixel: ", img[1, 1])
print("bottom-right pixel: ", img[2, 2])
print(img)

# center pixel to 255
# top-left pixel to 0
# bottom-right pixel to 128
img[1, 1] = 255
img[0, 0] = 0
img[2, 2] = 128
print("After changements")
print(img)
```

```python
# gives an ordered array with numbers from 0 to 99 similar to range in python
img = np.arange(100)
img = img.reshape((10, 10)).astype(np.uint8)
print(img)

# rows 2 to 5
# columns 3 to 7
crop = img[2:6, 3:8]
print("Crop shape: ", crop.shape)
print("Dtype: ", crop.dtype)
```

```python
# creating a black image
black_image = np.zeros((10, 10), dtype=np.uint8)

# draw a white square in the middle
black_image[2:8, 2:8] = 255

print(black_image)
```

```python
# basic operations:
img = np.array([
    [100, 12, 226],
    [150, 200, 250]
], dtype=np.uint8)

brighter_img = img + 30
safer_brigther_version = np.clip(img.astype(np.int16) + 30, 0, 255).astype(np.uint8)
darker = np.clip(img.astype(np.int16) - 80, 0, 255).astype(np.uint8)

print("---- original --------")
print(img)
print("---- dangerous brighter image --------")
print(brighter_img)
print("---- safer brighter image --------")
print(safer_brigther_version)
print("---- safer darker image --------")
print(darker)
```

```python
# increase contrast

img = np.array([
    [100, 12, 226],
    [150, 200, 250]
], dtype=np.uint8)

contrasted_img = np.clip(img.astype(np.float32) * 1.2, 0, 255).astype(np.uint8)

print(contrasted_img)
```

```python
img = np.zeros((3, 3, 3), dtype=np.uint8)
# Blue channel = 255
# Green channel = 100
# Red channel = 50

img[:, :, 0] = 255
img[:, :, 1] = 100
img[:, :, 2] = 50

print(img)

# print one pixel
print(img[0, 0])

blue = img[:, :, 0]
# every channel has the dimension of the image
print("Blue shape: ", blue.shape) 
```

```python
# binary mask

img = np.array([
    [10, 80, 120],
    [30, 150, 220],
    [5, 90, 255]
], dtype=np.uint8)

mask = img > 100

print("Created mask")
print(mask)
print("An image full of zeros with the same shape an dtype that other image")
result = np.zeros_like(img)
print(result)

# access only pixels whose index match the ones set in True inside the mask
result[mask] = img[mask]
print("Image filtered with the mask")
print(result)
```

```python
# threshold

img = np.array([
    [10, 80, 120],
    [30, 150, 220],
    [5, 90, 255]
], dtype=np.uint8)

binary = np.where(img > 100, 255, 0).astype(np.uint8)
print(binary)
```

```python
# Draw with NumPy
# draw a white square
# draw a gray horizontal line
# draw a gray vertical line

import matplotlib.pyplot as plt
import cv2

gray_scale_image = np.zeros((100, 100), dtype=np.uint8)
# drawing square
gray_scale_image[30:70, 30:70] = 255
# drawing horizontal line
gray_scale_image[15, 30:70] = 180
# drawing vertical line
gray_scale_image[30:70, 84] = 100

plt.imshow(gray_scale_image, cmap='gray')
plt.axis('off')
plt.show()

#cv2.imshow("Numpy draw", gray_scale_image)
#cv2.waitKey(0)
#cvé.destroyAllWindows()
```

```python
# Simple fake image filter
# brighter version
# darker version
# higher contrast version
# binary threshold version
# inverted version

img = np.tile(np.arange(0, 256), (256, 1)).astype(np.uint8)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()

print("Image brigther")
brigther = np.clip(img.astype(np.int16) + 100, 0, 255).astype(np.uint8)
plt.imshow(brigther, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()


print("Image darker")
darker = np.clip(img.astype(np.int16) - 100, 0, 255).astype(np.uint8)
plt.imshow(darker, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()


print("Image with contrast")
contrasted = np.clip(img.astype(np.float32) * 1.2, 0, 255).astype(np.uint8)
plt.imshow(contrasted, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()

print("Binary image")
binary = np.where(img > 127, 255, 0)
plt.imshow(binary, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()


print("Inverted image")
inverted = 255 - img
plt.imshow(inverted, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```
